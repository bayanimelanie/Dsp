```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Define the unit sample sequence (delta function)
def unit_sample_sequence(n_range, n0):
    """
    Generates a unit sample sequence.

    Parameters:
    - n_range: Range of n values (list or array)
    - n0: Position where the impulse occurs (typically 0)

    Returns:
    - delta: The unit sample sequence
    """
    delta = np.zeros(len(n_range))
    for i, n in enumerate(n_range):
        if n == n0:
            delta[i] = 1
    return delta

# Define the range of n
n_range = np.arange(-10, 11)  # From n=-10 to n=10

# Position of the impulse at n=0
n0 = 0

# Generate the unit sample sequence
delta = unit_sample_sequence(n_range, n0)

# Plot the unit sample sequence
plt.stem(n_range, delta, use_line_collection=True)
plt.title('Unit Sample Sequence (Delta Function)')
plt.xlabel('n')
plt.ylabel('Î´[n]')
plt.grid(True)
plt.show()
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Define the unit step function
def unit_step_sequence(n_range, n0):
    """
    Generates a unit step sequence.

    Parameters:
    - n_range: Range of n values (list or array)
    - n0: Step start position (typically 0)

    Returns:
    - step: The unit step sequence
    """
    step = np.zeros(len(n_range))
    for i, n in enumerate(n_range):
        if n >= n0:
            step[i] = 1
    return step

# Define the range of n
n_range = np.arange(-10, 11)  # From n=-10 to n=10

# Position where the step starts (usually at n=0)
n0 = 0

# Generate the unit step sequence
step = unit_step_sequence(n_range, n0)

# Plot the unit step sequence
plt.stem(n_range, step, use_line_collection=True)
plt.title('Unit Step Sequence (u[n])')
plt.xlabel('n')
plt.ylabel('u[n]')
plt.grid(True)
plt.show()
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Define the unit ramp function
def unit_ramp_sequence(n_range, n0):
    """
    Generates a unit ramp sequence.

    Parameters:
    - n_range: Range of n values (list or array)
    - n0: Position where ramp starts (typically 0)

    Returns:
    - ramp: The unit ramp sequence
    """
    ramp = np.zeros(len(n_range))
    for i, n in enumerate(n_range):
        if n >= n0:
            ramp[i] = n
    return ramp

# Define the range of n
n_range = np.arange(-10, 11)  # From n=-10 to n=10

# Position where the ramp starts (usually at n=0)
n0 = 0

# Generate the unit ramp sequence
ramp = unit_ramp_sequence(n_range, n0)

# Plot the unit ramp sequence
plt.stem(n_range, ramp, use_line_collection=True)
plt.title('Unit Ramp Sequence (r[n])')
plt.xlabel('n')
plt.ylabel('r[n]')
plt.grid(True)
plt.show()
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Define the exponential signal function
def exponential_signal(n_range, A, alpha):
    """
    Generates an exponential signal.

    Parameters:
    - n_range: Range of n values (list or array)
    - A: Amplitude of the signal
    - alpha: Exponential growth/decay rate

    Returns:
    - exp_signal: The exponential signal
    """
    exp_signal = A * np.exp(alpha * n_range)
    return exp_signal

# Define the range of n (discrete-time indices)
n_range = np.arange(-10, 11)  # From n=-10 to n=10

# Amplitude of the signal (A)
A = 1

# Exponential growth/decay rate (alpha)
alpha = 0.2  # Positive for growth, negative for decay

# Generate the exponential signal
exp_signal = exponential_signal(n_range, A, alpha)

# Plot the exponential signal
plt.stem(n_range, exp_signal, use_line_collection=True)
plt.title('Exponential Signal')
plt.xlabel('n')
plt.ylabel('x[n] = A * exp(Î± * n)')
plt.grid(True)
plt.show()
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Define the square wave function
def square_wave(n_range, period):
    """
    Generates a square wave signal.

    Parameters:
    - n_range: Range of n values (list or array)
    - period: Period of the square wave

    Returns:
    - sq_wave: The square wave signal
    """
    sq_wave = np.sign(np.sin(2 * np.pi * n_range / period))
    return sq_wave

# Define the range of n (discrete-time indices)
n_range = np.arange(0, 50)  # From n=0 to n=49

# Period of the square wave
period = 10  # Change this to adjust the frequency of the square wave

# Generate the square wave signal
sq_wave = square_wave(n_range, period)

# Plot the square wave signal
plt.stem(n_range, sq_wave, use_line_collection=True)
plt.title('Square Wave Signal')
plt.xlabel('n')
plt.ylabel('x[n]')
plt.grid(True)
plt.show()
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Define the triangular wave function
def triangular_wave(n_range, period, amplitude=1):
    """
    Generates a triangular wave signal.

    Parameters:
    - n_range: Range of n values (list or array)
    - period: Period of the triangular wave
    - amplitude: Peak amplitude of the wave

    Returns:
    - tri_wave: The triangular wave signal
    """
    # Generate a sawtooth wave with a width of 0.5 to make it symmetric like a triangle
    tri_wave = amplitude * (2 * np.abs(2 * (n_range / period - np.floor(0.5 + n_range / period))) - 1)
    return tri_wave

# Define the range of n (discrete-time indices)
n_range = np.arange(0, 50)  # From n=0 to n=49

# Period of the triangular wave
period = 10  # Adjust to control the frequency

# Amplitude of the triangular wave
amplitude = 1

# Generate the triangular wave signal
tri_wave = triangular_wave(n_range, period, amplitude)

# Plot the triangular wave signal
plt.stem(n_range, tri_wave, use_line_collection=True)
plt.title('Triangular Wave Signal')
plt.xlabel('n')
plt.ylabel('x[n]')
plt.grid(True)
plt.show()
```
